# 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = './malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df=pd.read_excel('./시도별 전출입 인구수.xlsx', engine='openpyxl', header=0)
df
```

```python
df.shape
```

```python
df=df.fillna(method='ffill')
```

```python
df
```

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
df_seoul=df[mask]

df_seoul
df_seoul=df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index(['전입지'], inplace=True)
```

```python
sr_one = df_seoul.loc['경기도']
sr_one
```

```python
plt.plot(sr_one.index, sr_one.values)
```

## 차트 제목, 축 이름 추가

```python
plt.plot(sr_one)

plt.title('서울 -> 경기 인구 이동')

plt.xlabel('기간')
plt.ylabel('이동인구수')

plt.show()
```

### 한글 폰트 관련 : 맨 윗부분 참고

## 그래프 꾸미기

```python
plt.style.use('ggplot')

plt.figure(figsize=(20,8)) 

# plt.xticks(rotation='vertical') # vertical 대신 각도를 숫자로 지정해도 됨
plt.xticks(rotation=-70)
# ↑↑↑↑↑↑↑↑↑ 전체적인 부분은 plot 설정보다 앞에 있어야함

plt.plot(sr_one.index, sr_one.values, marker='o', markersize=10) # plot()의 매개변수로 그래프 디자인을 바꿀 수 있음

plt.title('서울 -> 경기 인구 이동',size=30)

plt.xlabel('기간',size=20)
plt.ylabel('이동인구수',size=20)

plt.legend(labels=['서울 -> 경기'], loc='best', fontsize=15)

plt.show()

```

## 화면 분할하여 여러개 그래프 그리기

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)


ax1.plot(sr_one,'o',markersize=10)
ax2.plot(sr_one,marker='o',markerfacecolor='green',markersize=10,color='olive',linewidth=2,label='서울 -> 경기')
ax2.legend(loc='best')


ax1.set_ylim(50000,800000)
ax2.set_ylim(50000,800000)

ax2.set_title('서울->경기 인구 이동')

ax1.set_xticklabels(sr_one.index, rotation=75)
ax2.set_xticklabels(sr_one.index, rotation=75)


plt.show()
```

```python
fig=plt.figure(figsize=(20,5))
ax=fig.add_subplot(1,1,1)

ax.plot(sr_one,marker='o',markerfacecolor='yellow',markersize=10,color='red',linewidth=2,label='서울 -> 경기')
ax.legend(loc='best')


ax.set_ylim(50000,800000)

ax.set_title('서울->경기 인구 이동')

ax.set_xticklabels(sr_one.index, rotation=75)

ax.tick_params(axis='x',labelsize=11)

plt.show()
```

```python
col_years=list(map(str,range(1970,2017+1)))
df_3=df_seoul.loc[['충청남도','경상북도','강원도'],col_years]

plt.style.use('ggplot')

fig=plt.figure(figsize=(20,5))
ax=fig.add_subplot(1,1,1)

ax.plot(col_years, df_3.loc['충청남도',:], marker='o',markerfacecolor='yellow', markersize=10,color='green',linewidth=2,label='서울 -> 충남')
ax.plot(col_years, df_3.loc['경상북도',:], marker='o',markerfacecolor='violet', markersize=10,color='red',linewidth=2,label='서울 -> 경북')
ax.plot(col_years, df_3.loc['강원도',:], marker='o',markerfacecolor='lightblue', markersize=10,color='navy',linewidth=2,label='서울 -> 강원')

ax.legend(loc='best')

ax.set_title('서울 -> 충남, 경북, 강원 인구이동', size=21)

ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)

ax.set_xticklabels(col_years, rotation=70)

ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```

```python
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]

plt.style.use('bmh')


fig=plt.figure(figsize=(20,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

ax1.plot(col_years, df_4.loc['충청남도',:], marker='o',markerfacecolor='olive', markersize=7,color='green',linewidth=2,label='서울 -> 충남')
ax2.plot(col_years, df_4.loc['경상북도',:], marker='s',markerfacecolor='tomato', markersize=7,color='red',linewidth=2,label='서울 -> 경북')
ax3.plot(col_years, df_4.loc['강원도',:], marker='*',markerfacecolor='orange', markersize=10,color='coral',linewidth=2,label='서울 -> 강원')
ax4.plot(col_years, df_4.loc['전라남도',:], marker='D',markerfacecolor='blue', markersize=6,color='navy',linewidth=2,label='서울 -> 전남')

ax1.legend(loc='best')
ax2.legend(loc='upper right')
ax3.legend(loc='lower right')
ax4.legend(loc='best')

ax1.set_ylim(5000,60000)
ax2.set_ylim(5000,60000)
ax3.set_ylim(5000,60000)
ax4.set_ylim(5000,60000)

ax1.set_title('서울 -> 충남 인구이동', size=15)
ax2.set_title('서울 -> 경북 인구이동', size=15)
ax3.set_title('서울 -> 강원 인구이동', size=15)
ax4.set_title('서울 -> 전남 인구이동', size=15)

ax1.set_xticklabels(col_years, rotation=65)
ax2.set_xticklabels(col_years, rotation=65)
ax3.set_xticklabels(col_years, rotation=65)
ax4.set_xticklabels(col_years, rotation=65)



plt.show()
```

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul=df[mask]

df_seoul
df_seoul=df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index(['전입지'], inplace=True)
```

```python
col_years=list(map(str,range(1970,2017+1)))
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]
df_4=df_4.transpose()

plt.style.use('bmh')


df_4.index = df_4.index.map(int)

df_4.plot(kind='area', stacked=False, alpha=0.1, figsize=(20,10))

plt.ylim(5000, 60000)
plt.xlim(1970, 2017)
plt.title('서울 -> 타시도 인구이동', size=30)
plt.ylabel('이동 인구 수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best',fontsize=15)

plt.show()
```

```python
col_years=list(map(str,range(1970,2017+1)))
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]
df_4=df_4.transpose()

plt.style.use('fast')

df_4.index = df_4.index.map(int)



ax=df_4.plot(kind='area', stacked=True, alpha=0.8, figsize=(20,10))

ax.set_ylim(5000, 200000)
ax.set_xlim(1970, 2017)

ax.set_title('서울 -> 타시도 인구이동', size=30)
ax.set_ylabel('이동 인구 수', size=20)
ax.set_xlabel('기간', size=20)
ax.legend(loc='best',fontsize=15)

plt.show()
```

```python
col_years=list(map(str,range(2010,2017+1)))
df_5 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]
df_5=df_5.transpose()


plt.style.use('bmh')

df_5.index = df_5.index.map(int)

df_5.plot(kind='bar', figsize=(20,10), width=0.7, color=['coral','goldenrod','steelblue','blueviolet'])

plt.title('서울 -> 타시도 인구이동', size=30)
plt.ylabel('이동 인구 수', size=20)
plt.xlabel('기간', size=20)
plt.ylim(5000, 30000)
plt.legend(loc='best',fontsize=15)

plt.show()
```

```python
col_years=list(map(str,range(2010,2017+1)))
df_6 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]

df_6['합계']=df_6.sum(axis=1)

df_total=df_6[['합계']].sort_values(by='합계', ascending=True)

plt.style.use('fast')

df_total.plot(kind='barh',color='steelblue', width=0.7, figsize=(10,5) )

plt.title('서울 -> 타시도 인구이동', size=30)
plt.ylabel('전입지', size=20)
plt.xlabel('이동 인구수', size=20)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
font_path = './malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.style.use('default')
plt.rcParams['axes.unicode_minus']=False
```

```python
df=pd.read_csv('./auto-mpg.csv', header=None)
```

```python
df.columns = ['mpg','cylinders','displacement','horsepower', 'weight', 'acceleration', 'model year', 'origin', 'name']
df
```

```python
df['mpg'].plot(kind='hist',bins=10,color='steelblue',figsize=(10,5))

plt.title('Histogram')
plt.xlabel('mpg')
plt.show()
```

```python
df.plot(kind='scatter', x='weight', y='mpg',c='coral', s=10, figsize=(10,5))
plt.title('Scatter Plot - mpg vs weight')
plt.show()
```

```python
cylinders_size=df.cylinders/df.cylinders.max() *500
```

```python
df.plot(kind='scatter', x='weight', y='mpg',c='coral', figsize=(10,5), s=cylinders_size, alpha=0.2)
plt.title('Scatter Plot - mpg - weight - cylinders')

plt.show()
```

```python
df.plot(kind='scatter', x='weight', y='mpg', marker='+', 
         figsize=(13,5), cmap='winter', c=cylinders_size, s=80, alpha=0.3)
plt.title('Scatter Plot - mpg - weight - cylinders')

plt.savefig('./scatter.png')
plt.savefig('./scatter_transparent.png', transparent=True)

plt.show()
```

```python
df['count'] = 1
df_origin=df.groupby('origin').sum()
print(df_origin.head())
```

```python
df_origin.index=['USA', 'EU', 'JPN']

df_origin['count'].plot(kind='pie', figsize=(6,5), fontsize=10, autopct='%1.1f%%',
                        startangle=0)

plt.title('Model Origin', size=20)
plt.axis('equal')
plt.legend(labels=df_origin.index, loc='upper right')
plt.show()
```

```python
plt.rc('font', family='Malgun Gothic')
fig=plt.figure(figsize=(15,5))

ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

ax1.boxplot(x=[df[df['origin']==1]['mpg'],
              df[df['origin']==2]['mpg'],
              df[df['origin']==3]['mpg']],
            labels=['USA','EU','JPN'])

ax2.boxplot(x=[df[df['origin']==1]['mpg'],
              df[df['origin']==2]['mpg'],
              df[df['origin']==3]['mpg']],
            labels=['USA','EU','JPN'],
            vert=False)

ax1.set_title('제조국가별 연비 분포(수직 박스 플록)')
ax2.set_title('제조국가별 연비 분포(수평 박스 플록)')

plt.show()
```

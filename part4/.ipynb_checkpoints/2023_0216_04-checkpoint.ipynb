```python
import pandas as pd
import folium
```

```python
df=pd.read_excel('./서울지역 대학교 위치.xlsx', engine='openpyxl')

print(df)

df=df.rename(columns={'Unnamed: 0': '대학명'})
df=df.set_index('대학명')
```

```python
seoul_map = folium.Map(location=[37.55,126.98], zoom_start=15)

for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat,lng], popup=folium.Popup(name, max_width=500)).add_to(seoul_map)

seoul_map
```

```python
seoul_map2 = folium.Map(location=[37.55,126.98], zoom_start=15)

for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat,lng], radius=10, color='brown', 
                        fill=True, fill_color='olive', fill_opacity=0.8, 
                        popup=folium.Popup(name, max_width=500)).add_to(seoul_map2)
    
seoul_map2
```
